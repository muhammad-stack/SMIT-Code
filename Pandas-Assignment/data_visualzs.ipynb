## 1. Load and Display Data â†’ Read the dataset and print the first 5 rows

```python
import pandas as pd

file = pd.read_csv('covid_19_country_wise_latest-Copy1.csv')

file.head()
```

## 2. Set the Dataset index is Unnamed: 0


```python
file.set_index('Unnamed: 0',inplace = True)
```

```python
file
```

## 3. Now replace the index column Unnamed: 0 name with index

```python
file.index.name = 'Index'
```

```python
file
```

## 4. Create Function that take dataset columns and replace whitespace
## with Underscore and update dataset automatically and the pass the
## dataset.

```python
def replace_whitespace_with_underscore(file):
    file.columns = file.columns.str.replace(' ', '_')
    return file
```

```python
replace_whitespace_with_underscore(file)
```

## 5. Check Basic Information â†’ Find column names, data types, and missing values.

```python
file.info()
```

## Task 6: Fill NaN Values with the Mean of the Column



```python
# Select the columns having the datatype float
numerical_cols = file.select_dtypes(include=['float64']).columns

# Fill that columns with there respective mean
file[numerical_cols] = file[numerical_cols].fillna(file[numerical_cols].mean())
```

```python
file
```

## 7. Count the number of unique countries in the dataset.

```python
file['Country/Region'].nunique()
```

## Step 8: Check and Remove Duplicate Country Entries



```python
duplicates = file['Country/Region'].duplicated().sum()
print(duplicates)

if duplicates > 0:
    file = file.drop_duplicates(subset=['Country/Region'])
```

## Step 9: Calculate Mean, Median, and Standard Deviation of Total Cases



```python
mean_cases = file['Confirmed'].mean()
median_cases = file['Confirmed'].median()
std_cases = file['Confirmed'].std()
print(f"Mean Total Cases: {mean_cases}")
print(f"Median Total Cases: {median_cases}")
print(f"Standard Deviation of Total Cases: {std_cases}")
```

## Step 10: Replace Strings in 'Deaths' Column with Mean



```python

file['Deaths'] = pd.to_numeric(file['Deaths'], errors='coerce')

# Calculate mean and fill NaN
mean_deaths = file['Deaths'].mean()
file['Deaths'] = file['Deaths'].fillna(mean_deaths)
```

```python
file.Deaths
```

## Step 11: Change Datatype of 'Deaths' Column



```python
file['Deaths'] = file['Deaths'].astype(float)
```

```python
file.Deaths.dtype
```

## Step 12: Calculate Total Deaths and Recoveries Worldwide



```python
total_deaths = file['Deaths'].sum()
total_recoveries = file['Recovered'].sum()
print(f"Total Deaths Worldwide: {total_deaths}")
print(f"Total Recoveries Worldwide: {total_recoveries}")
```

## Step 13: Count Countries with Over 1 Million Total Case



```python
million_cases = (file['Confirmed'] > 1000000).sum()
print(f"Countries with >1M Total Cases: {million_cases}")
```

## Step 14: Find Countries with Recovery Rate Above 95%



```python
recovery_rate = file['Recovered'] / file['Confirmed']
high_recovery_countries = file[recovery_rate > 0.95]['Country/Region'].tolist()
print(f"Countries with >95% Recovery Rate: {high_recovery_countries}")
```

## Step 15: Drop 'WHO_Region' and 'Confirmed' Columns



```python
file = file.drop(['WHO_Region', 'Confirmed'], axis=1)
```

## Step 16: Find Country with Maximum Deaths



```python
max_deaths_country = file.loc[file['Deaths'].idxmax(), 'Country/Region']
print(f"Country with Maximum Deaths: {max_deaths_country}")
```

## Step 17: Sort Countries by Deaths in Descending Order



```python
file = file.sort_values('Deaths', ascending=False)
print(file[['Country/Region', 'Deaths']].head())
```

## Step 18: Create 'Total_Intellectuals' Column



```python
file['Total_Intellectuals'] = file['Deaths'] + file['Recovered'] + file['Active']
```

```python
file['Total_Intellectuals']
```

## Step 19: Calculate Death Rate



```python
file['Death_Rate'] = file['Deaths'] / file['Total_Intellectuals']
```

```python
file['Death_Rate'].sum()
```

## Step 20: Identify Countries with Increasing Cases and Low Death Rates



```python
increasing_low_death = file[(file['New_cases'] > 0) & (file['Death_Rate'] < 0.01)]['Country/Region'].tolist()
print(f"Countries with Increasing Cases and Low Death Rates: {increasing_low_death}")
```

## Step 21: Print Head of 'Country/Region' and 'Death_Rate'



```python
file[['Death_Rate' ,'Country/Region']].head()
```

## Step 22: Create a Scatter Plot of Total Cases vs. Total Deaths



```python
import matplotlib.pyplot as plt

plt.scatter(file['Total_Intellectuals'], file['Deaths'])
plt.xlabel('Total Cases')
plt.ylabel('Total Deaths')
plt.title('Total Cases vs. Total Deaths')
plt.show()
```

## Step 23: Save the Dataset to CSV



```python
file.to_csv('modified_dataset.csv', index=True )
```

```python
import numpy as np
def transform_year(year):
    if year.isalpha():
        return np.nan
    else:
        if year.isnumeric():
            return int(year)
```

```python
data.year = data.year.apply(transform_year)
```

```python
data.year = data.year.fillna(method='ffill')
```

```python
employment_data = {'AgeGroup':['18-25', '25-35','45-55','55-65',"65+"],
                   "Employed%":[85,95,97,97,100]
                  }
```

```python
data = pd.DataFrame(employment_data)
data['Total'] =100
data['Unemployed%'] = data.Total - data['Employed%']
```

```python
data = data[['AgeGroup', 'Employed%','Unemployed%','Total' ]]
```

```python
data
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 4))
bar_width = 0.4
x = range(len(data))

plt.bar(x, data["Employed%"], width=bar_width, label='Employed', color='blue')
plt.bar([i + bar_width for i in x], data["Unemployed%"], width=bar_width, label='Unemployed', color='orange')

plt.xlabel("Agegroup")
plt.ylabel("Number of People")
plt.title("Employment Distribution by Age Group")
plt.xticks([i + bar_width / 2 for i in x], data["AgeGroup"])
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

plt.show()
```

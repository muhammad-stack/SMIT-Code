```python
import pandas as pd`
```

```python
state0 = pd.read_csv("states0.csv").drop("Unnamed: 0",axis=1)
```

```python
state0
```

```python
state0.Hispanic = state0.Hispanic.str.replace("%", "").astype(float)*100
```

```python
state0.White = state0.White.str.replace("%", "").astype(float)*100
state0.Black = state0.Black.str.replace("%", "").astype(float)*100
state0.Native = state0.Native.str.replace("%", "").astype(float)*100
state0.Asian = state0.Asian.str.replace("%", "").astype(float)*100
state0.Pacific = state0.Pacific.str.replace("%", "").astype(float)*100
```

```python
state0.Income = state0.Income.str.replace("$", "").astype(float)
```

```python
state0
```

```python
state0["males"],state0['females']=state0.GenderPop.str.split("_").str[0],state0.GenderPop.str.split("_").str[1]
```

```python
state0['males']=state0['males'].str.replace("M","").astype(int)
state0['females']=state0['females'].str.replace("F","").astype(int)
```

```python
del state0['GenderPop']
```

```python
state0
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x=state0.Income, y=state0.females)
plt.xlabel('Income')
plt.ylabel('Females')
plt.title('average income in a state vs proportion of women in that state')
```
